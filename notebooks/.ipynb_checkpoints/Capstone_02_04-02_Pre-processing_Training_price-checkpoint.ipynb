# Springboard Capstone 02 - Energy generation prediction based on energy and weather data
## Pre-processing

```python
#Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from dateutil import tz
from pytz import timezone
from pandas import Grouper
from pandas import concat
from pandas import DataFrame
import matplotlib.ticker as mtick
```

```python
#import dataframe and deal with date column
#eng_wea_df = pd.read_csv('data/energy_weather.csv')
eng_wea_df = pd.read_pickle('data/energy_weather_shift.pkl')

eng_wea_df['date'] = pd.to_datetime(eng_wea_df['date'], utc=False)
#Convert to Madrid timezone
eng_wea_df['date'] = pd.to_datetime(eng_wea_df['date']).dt.date

# #eng_wea_df.set_index('date', inplace=True)
# energy_weather_df.drop(columns='Unnamed: 0', inplace=True)

eng_wea_df.head()
```

```python
eng_wea_df.columns
```

```python
# eng_wea_df.columns = eng_wea_df.columns.get_level_values(0)
# eng_wea_df.head(15)
```

```python
eng_wea_df.set_index('date', inplace=True)
```

```python
eng_wea_df.info()
```

## 2.0 Converting Categorical Features

```python
eng_wea_df = pd.get_dummies(eng_wea_df, columns=['city_name', 'month', 'day_of_week'])
eng_wea_df.info()
```

## 3.0 Splitting data

```python
from sklearn.model_selection import train_test_split

X = eng_wea_df.drop(columns=['price actual_daily_max', 'price day ahead_daily_max'])
y = eng_wea_df['price actual_daily_max']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

## 4.0 Pipeline

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import cross_validate, GridSearchCV, learning_curve
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
pipe = make_pipeline(
    SimpleImputer(strategy='median'), 
    StandardScaler(),
    LinearRegression()
)
```

```python
import timeit
start = timeit.default_timer()
print("The start time is :", start)
pipe.fit(X_train, y_train)
print("The difference of time is :", 
              timeit.default_timer() - start)
```

```python
y_tr_pred = pipe.predict(X_train)
y_te_pred = pipe.predict(X_test)
```

```python
r2_score(y_train, y_tr_pred), r2_score(y_test, y_te_pred)
```

```python
mean_absolute_error(y_train, y_tr_pred), mean_absolute_error(y_test, y_te_pred)
```

```python
mean_squared_error(y_train, y_tr_pred), mean_squared_error(y_test, y_te_pred)
```

```python
import numpy as np
#root mean_squared error
np.sqrt(mean_squared_error(y_train, y_tr_pred)), np.sqrt(mean_squared_error(y_test, y_te_pred))
```

## Adding SelectKBest using default k of 10

```python
pipe = make_pipeline(
    SimpleImputer(strategy='median'), 
    StandardScaler(),
    SelectKBest(f_regression),
    LinearRegression()
)
```

```python
start = timeit.default_timer()
print("The start time is :", start)
pipe.fit(X_train, y_train)
print("The difference of time is :", 
              timeit.default_timer() - start)
```

```python
y_tr_pred = pipe.predict(X_train)
y_te_pred = pipe.predict(X_test)
```

```python
r2_score(y_train, y_tr_pred), r2_score(y_test, y_te_pred)
```

```python
mean_absolute_error(y_train, y_tr_pred), mean_absolute_error(y_test, y_te_pred)
```

```python
mean_squared_error(y_train, y_tr_pred), mean_squared_error(y_test, y_te_pred)
```

```python
#root mean_squared error
np.sqrt(mean_squared_error(y_train, y_tr_pred)), np.sqrt(mean_squared_error(y_test, y_te_pred))
```

## Cross validation

```python
start = timeit.default_timer()
print("The start time is :", start)
cv_results = cross_validate(pipe, X_train, y_train, cv=5)
print("The difference of time is :", 
              timeit.default_timer() - start)
```

```python
cv_scores = cv_results['test_score']
cv_scores
```

```python
np.mean(cv_scores), np.std(cv_scores)
```

```python
np.round((np.mean(cv_scores) - 2 * np.std(cv_scores), np.mean(cv_scores) + 2 * np.std(cv_scores)), 2)
```

## Hyperparameter Search with GridSearchCV

```python
k = [k+1 for k in range(len(X_train.columns))]
grid_params = {'selectkbest__k': k}
grid_params
```

```python
lr_grid_cv = GridSearchCV(pipe, param_grid=grid_params, cv=5, n_jobs=-1)
```

```python
start = timeit.default_timer()
print("The start time is :", start)

lr_grid_cv.fit(X_train, y_train)

print("The difference of time is :", 
              timeit.default_timer() - start)
```

```python
score_mean = lr_grid_cv.cv_results_['mean_test_score']
score_std = lr_grid_cv.cv_results_['std_test_score']
cv_k = [k for k in lr_grid_cv.cv_results_['param_selectkbest__k']]
```

```python
lr_grid_cv.best_params_
```

```python
best_k = lr_grid_cv.best_params_['selectkbest__k']
plt.subplots(figsize=(10, 5))
plt.errorbar(cv_k, score_mean, yerr=score_std)
plt.axvline(x=best_k, c='r', ls='--', alpha=.5)
plt.xlabel('k')
plt.ylabel('CV score (r-squared)')
plt.title('Pipeline mean CV score (error bars +/- 1sd)');

plt.savefig('images/price_linarCV_bestk.png')
```

```python
selected = lr_grid_cv.best_estimator_.named_steps.selectkbest.get_support()
```

```python
coefs = lr_grid_cv.best_estimator_.named_steps.linearregression.coef_
features = X_train.columns[selected]
pd.Series(coefs, index=features).sort_values(ascending=False)
```

```python
plt.subplots(figsize=(10, 5))
coefs = lr_grid_cv.best_estimator_.named_steps.linearregression.coef_
features = X_train.columns[selected]
lr_feat_coef = pd.Series(coefs, index=features).sort_values(ascending=False)
lr_feat_coef.plot(kind='bar')
plt.xlabel('features')
plt.ylabel('coefficients')
plt.title('Best linear regressor feature coefficients');

plt.savefig('images/price_best_linreg_featurecoef.png', bbox_inches='tight')
```

## Random Forest

```python
RF_pipe = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler(),
    RandomForestRegressor(random_state=47)
)
```

```python
start = timeit.default_timer()
print("The start time is :", start)

rf_default_cv_results = cross_validate(RF_pipe, X_train, y_train, cv=5)

print("The difference of time is :", 
              timeit.default_timer() - start)
```

```python
rf_cv_scores = rf_default_cv_results['test_score']
rf_cv_scores
```

```python
np.mean(rf_cv_scores), np.std(rf_cv_scores)
```

## Hyperparameter search using GridSearchCV

```python
n_est = [int(n) for n in np.logspace(start=1, stop=3, num=5)]
grid_params = {
        'randomforestregressor__n_estimators': n_est,
        'standardscaler': [StandardScaler(), None],
        'simpleimputer__strategy': ['mean', 'median']
}
grid_params
```

```python
rf_grid_cv = GridSearchCV(RF_pipe, param_grid=grid_params, cv=5, n_jobs=-1)
```

```python
start = timeit.default_timer()
print("The start time is :", start)

rf_grid_cv.fit(X_train, y_train)

print("The difference of time is :", 
              timeit.default_timer() - start)
```

```python
rf_grid_cv.best_params_
```

```python
start = timeit.default_timer()
print("The start time is :", start)

rf_best_cv_results = cross_validate(rf_grid_cv.best_estimator_, X_train, y_train, cv=5)

print("The difference of time is :", 
              timeit.default_timer() - start)
```

```python
rf_best_scores = rf_best_cv_results['test_score']
rf_best_scores
```

```python
np.mean(rf_best_scores), np.std(rf_best_scores)
```

```python
plt.subplots(figsize=(10, 5))
imps = rf_grid_cv.best_estimator_.named_steps.randomforestregressor.feature_importances_
rf_feat_imps = pd.Series(imps, index=X_train.columns).sort_values(ascending=False)
rf_feat_imps.plot(kind='bar')
plt.xlabel('features')
plt.ylabel('importance')
plt.title('Best random forest regressor feature importances');

plt.savefig('images/price_best_RF_featureimp.png',  bbox_inches='tight')
```

## Selecting best model

```python
lr_mae_mean = np.mean(-1 * lr_neg_mae['test_score'])
lr_mae_std = np.std(-1 * lr_neg_mae['test_score'])
lr_mae_mean, lr_mae_std
```

```python
mean_absolute_error(y_test, lr_grid_cv.best_estimator_.predict(X_test))
```

```python
start = timeit.default_timer()
print("The start time is :", start)

rf_neg_mae = cross_validate(rf_grid_cv.best_estimator_, X_train, y_train, 
                            scoring='neg_mean_absolute_error', cv=5, n_jobs=-1)

print("The difference of time is :", 
              timeit.default_timer() - start)
```

```python
rf_mae_mean = np.mean(-1 * rf_neg_mae['test_score'])
rf_mae_std = np.std(-1 * rf_neg_mae['test_score'])
rf_mae_mean, rf_mae_std
```

## Further assessment

```python
fractions = [.2, .25, .3, .35, .4, .45, .5, .6, .75, .8, 1.0]
train_size, train_scores, test_scores = learning_curve(pipe, X_train, y_train, train_sizes=fractions)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
```

```python
plt.subplots(figsize=(10, 5))
plt.errorbar(train_size, test_scores_mean, yerr=test_scores_std)
plt.xlabel('Training set size')
plt.ylabel('CV scores')
plt.title('Cross-validation score as training set size increases');

plt.savefig('images/price_linear_CV_trainingsetsize.png')
```
