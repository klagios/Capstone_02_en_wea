# Springboard Capstone 02 - Energy price prediction based on weather
## Joining the data sets

```python
import pandas as pd
```

```python
energy_data = pd.read_csv('data/eng_grouped_max.csv')
weather_data = pd.read_csv('data/wea_grouped_max.csv')
```

```python
energy_data.head()
```

```python
energy_data.columns
```

```python
weather_data.head()
```

```python
weather_data.columns
```

```python
weather_data.rename(columns={'Unnamed: 0': 'date', 'Unnamed: 1': 'city_name'}, inplace=True)
weather_data.columns
```

```python
weather_data.drop(0, axis=0, inplace=True)
weather_data.columns
```

```python
weather_data.head()
```

```python
eng_weather = pd.merge(weather_data, energy_data)
eng_weather.head()
```

```python
eng_weather.head(15)
```

```python
eng_weather.to_csv('data/energy_weather.csv')
eng_weather.to_pickle('data/eng_weather.pkl')
```
