## Implement for entire df

```python
#create df subset
df_sub = energy_weather_max[['date', 'city_name', 'temp_daily_max', 'pressure_daily_max', 'humidity_daily_max', 'wind_speed_daily_max', 
                       'rain_1h_daily_max', 'rain_3h_daily_max', 'snow_3h_daily_max', 'clouds_all_daily_max']]
df_sub.head()
```

```python
df_sub.date = df_sub.date  + timedelta(days=1)
df_sub.head()
```

```python
energy_weather_shift = energy_weather_max.drop(columns=['temp_daily_max', 'pressure_daily_max', 'humidity_daily_max', 'wind_speed_daily_max', 
                       'rain_1h_daily_max', 'rain_3h_daily_max', 'snow_3h_daily_max', 'clouds_all_daily_max'])
```

```python
energy_weather_shift = pd.merge(energy_weather_shift, df_sub, how='left')
```

```python
#drop first date - don't have to drop last date if using left join
dropdate = pd.to_datetime("2015-01-01").date()
ind = energy_weather_shift[energy_weather_shift['date'] == dropdate].index
energy_weather_shift.drop(ind, axis=0, inplace=True)
energy_weather_shift.head()
```

```python
energy_weather_shift.reset_index(drop=True, inplace=True)
energy_weather_shift.head()
```
