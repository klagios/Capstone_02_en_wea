```python
#Import packagges
import pandas as pd
from datetime import timedelta
```

## 1.0 Import data<a id='1.0_import_data'></a>

```python
#Import data
energy_weather_max = pd.read_pickle("data/eng_weather.pkl")
```

```python
energy_weather_max.head(15)
```

```python
#energy_data['time'] = pd.to_datetime(energy_data['time'], utc=True)
import datetime

energy_weather_max['date'] = pd.to_datetime(energy_weather_max['date'], utc=False)
energy_weather_max['date'] = pd.to_datetime(energy_weather_max['date']).dt.date
#isinstance(mini_df['date'], datetime.date)
#type(mini_df['date'])
energy_weather_max.info()
```

## 2.0 Create a subset of the dataframe to test process on<a id='2.0_mini_df'></a>

```python
# test on mini_df
#cut_date = datetime.strptime('2015-01-04',  "%Y-%m-%d").date
enddate = pd.to_datetime("2015-01-04").date()
mini_df = energy_weather_max[energy_weather_max['date'] < enddate]
mini_df
```

```python
mini_df_sub = mini_df[['date', 'city_name', 'temp_daily_max', 'pressure_daily_max', 'humidity_daily_max', 'wind_speed_daily_max', 
                       'rain_1h_daily_max', 'rain_3h_daily_max', 'snow_3h_daily_max', 'clouds_all_daily_max']]
mini_df_sub.head()
```

### 2.1 Shift dates of mini dataframe<a id='2.1_shift_mini'></a>

```python
#Shift dates
mini_df_sub.date = mini_df_sub.date  + timedelta(days=1)
mini_df_sub.head()
```

```python
mini_df = mini_df.drop(columns=['temp_daily_max', 'pressure_daily_max', 'humidity_daily_max', 'wind_speed_daily_max', 
                       'rain_1h_daily_max', 'rain_3h_daily_max', 'snow_3h_daily_max', 'clouds_all_daily_max'])
```

```python
mini_df.head()
```

```python
mini_df_shift = pd.merge(mini_df, mini_df_sub, how='left')
```

```python
mini_df_shift
```

```python
#drop first date - don't have to drop last date if using left join
dropdate = pd.to_datetime("2015-01-01").date()
ind = mini_df[mini_df['date'] == dropdate].index
```

```python
mini_df.drop(ind, axis=0, inplace=True)
mini_df
```

## 3.0 Date shift entire dataframe<a id='3.0_shift_full'></a>

```python
#create df subset
df_sub = energy_weather_max[['date', 'city_name', 'temp_daily_max', 'pressure_daily_max', 'humidity_daily_max', 'wind_speed_daily_max', 
                       'rain_1h_daily_max', 'rain_3h_daily_max', 'snow_3h_daily_max', 'clouds_all_daily_max']]
df_sub.head()
```

```python
df_sub.date = df_sub.date  + timedelta(days=1)
df_sub.head()
```

```python
energy_weather_shift = energy_weather_max.drop(columns=['temp_daily_max', 'pressure_daily_max', 'humidity_daily_max', 'wind_speed_daily_max', 
                       'rain_1h_daily_max', 'rain_3h_daily_max', 'snow_3h_daily_max', 'clouds_all_daily_max'])
```

```python
energy_weather_shift = pd.merge(energy_weather_shift, df_sub, how='left')
```

```python
#drop first date - don't have to drop last date if using left join
dropdate = pd.to_datetime("2015-01-01").date()
ind = energy_weather_shift[energy_weather_shift['date'] == dropdate].index
energy_weather_shift.drop(ind, axis=0, inplace=True)
energy_weather_shift.head()
```

```python
energy_weather_shift.reset_index(drop=True, inplace=True)
energy_weather_shift.head()
```

## 4.0 Save dataframe<a id='4.0_save_df'></a>

```python
energy_weather_shift.to_pickle('data/energy_weather_shift.pkl')
```
